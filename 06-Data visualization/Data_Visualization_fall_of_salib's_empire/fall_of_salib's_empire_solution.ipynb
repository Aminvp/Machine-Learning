<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
سلیب یکی از بندگان خدا بود که متاسفانه به همه اعتماد داشت. اون رئیس بخش فروش یک شرکت بزرگ تولید جارو دستی بود. کارمندان از ساده لوحی او سو استفاده می‌کردند و کل روز را به بطالت می‌گذراندند. آنها روی صندلی سلیب یک سنسور نصب کرده بودند تا وقتی سلیب از صندلی خود بلند می‌شد، شَست کارمندان خبردار شده و تظاهر به کار سنگین بکنند. سلیب دید که مدتیست صادرات جارو دستی رو به وخامت می‌رود در حالی که کارمندان عزیز او سخت مشغول کارند. دوست زرنگ سلیب، مش عبدالله که میدانست کارمندان ذات پلیدی دارند، به سلیب گفت که یک ماه به جای کارمندان، تعدادی بز در اداره بگذارد و نتیجه را ببیند. سپس مقایسه‌ای بین فروش دوره حضور کارمندان و دوره حضور بزها قرار دهد و نتیجه را روی دیوار شرکت نصب کند تا کارمندان تحقیر شده و به خودشان بیایند. در این تمرین ما به سلیب کمک می‌کنیم تا نمودار میزان فروش جارو دستی در دوره را تولید کند.
</font>
</p>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
مت‌پلات به طور پیش‌فرض محدوده اعداد محور افقی و عمودی را از روی داده ورودی مشخص می‌کند. یعنی اگر داده ورودی در محور عمودی، بین ۱۰ تا ۲۰ باشد، مت‌پلات محدوده محور عمودی را بازه بین ۱۰ تا ۲۰ در نظر می‌گیرد. در مواردی لازم می‌شود که این بازه را تغییر دهیم و فرضا آن را بزرگ‌تر کنیم. برای این‌کار از متد‌های <code>set_ylim</code> و <code>set_xlim</code> استفاده می‌کنیم. 
<br>
وودی این تابع دو عدد، که ابتدا و انتهای محور را مشخص می‌کنند، می‌باشد. در این تمرین از این ویژگی مت‌پلات استفاده می‌کنیم.
</font>

برای رسم نمودار برای سلیب، عملیات زیر را انجام دهید:

<ul dir =rtl>

  <li>
در تابع <code>subplots</code> مقدار پارامتر <code>sharey</code> را برابر <code>True</code> قرار دهید تا اعداد محور افقی برای هر دو صفحه محور تکرار نشوند. (هر دو یکسان هستند و می‌توان با این روش نمودار خلوت‌تر و خوانا‌تری داشت.)</li>

  <li>
با استفاده از توابعی که برای تغییر محدوده محور افقی و عمودی در همین تمرین معرفی کردیم، محدوده محور افقی را برای هر دو نمودار تغییر داده و بازه ۱۵۰۰ تا ۳۵۰۰ را به این صفحه‌محورها اختصاص دهید.</li>
  <li>
عنوان صفحه‌محور سمت چپ را <code>farvardin</code> و عنوان صفحه‌محور سمت راست را <code>ordibehesht</code> قرار دهید.
</li>

<li>
عنوان محور افقی را برای هر دو صفحه‌محور برابر <code>days</code> و عنوان محور عمودی را برای صفحه‌محور سمت چپ برابر <code>count</code> قرار دهید. چون این محور را برای هر دو صفحه‌محور یکسان قرار دادیم، فقط محور عمودی صفحه‌محور سمت چپی نیاز به عنوان دارد.
</li>


<li>
از ویژگی <code>legend</code> مت‌پلات استفاده کرده و عنوان داده صفحه‌محور سمت چپ را <code>employees</code> و عنوان داده صفحه‌محور سمت راست را <code>goats</code> قرار دهید.
</li>

```python
import matplotlib.pyplot as plt

employees = [2883, 2944, 2774, 2343, 2239, 2063, 2329, 2652, 2500, 2502, 2684,
       2823, 2644, 2433, 2218, 2194, 2900, 2602, 2771, 2205, 2503, 2650,
       2639, 2996, 2296, 2231, 2026, 2658, 2187, 2869]

goats = [2958, 2425, 2360, 2039, 2909, 2206, 2359, 2198, 2619, 2649, 2076,
       2623, 2275, 2481, 2458, 2724, 2520, 2645, 2561, 2518, 2895, 2726,
       2500, 2578, 2542, 2585, 2931, 2564, 2420, 2930]

fig, ax = plt.subplots(1,2, figsize=(10,5), sharey=True)
ax[0].plot(employees,label = 'employees')
ax[0].set_ylim(1500,3500)
ax[0].set_title('farvardin')
ax[0].legend()
ax[0].set_ylabel('count')
ax[0].set_xlabel('days')

ax[1].plot(goats,label = 'goats')
ax[1].set_ylim(1500,3500)
ax[1].set_title('ordibehesht')
ax[1].legend()
ax[1].set_xlabel('days');
```

```python
import zipfile
import joblib
import json

joblib.dump(fig,"fig")

def compress(file_names):
    print("File Paths:")
    print(file_names)
    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED
    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ["fig","fall_of_salib's_empire_solution.ipynb"]
compress(file_names)
```
