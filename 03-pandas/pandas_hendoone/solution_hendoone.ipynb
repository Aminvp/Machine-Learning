```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('./food.csv')
df
```

<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
        با استفاده از تابع value_counts تعداد تکرار عناصر منحصر به فرد ستون flavor_profile را می‌شماریم.
    </font>
</div>


```python
counts = df.flavor_profile.value_counts()
counts
```

<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
        نرخ غذاهای ترش برابر تعداد غذاهای ترش بر مجموع تعداد همه غذاهاست.
        <br>
        عبارت <code>counts.sum()</code> همه مقادیر متغیر counts که در سلول بالا نشان داده‌شده است را جمع می‌کند.
        
</div>


```python
spicy_ratio = counts['spicy'] / counts.sum()
spicy_ratio
```

<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
        تابع <code>calculate_full_time</code> یک سری (ستون) می‌گیرد و جمع مقادیری را محسابه می‌کند که کلیدشان prep_time و cook_time باشد.
</div>


```python
def calculate_full_time (row) :
    return row['prep_time'] + row['cook_time']
```

```python
df['full_time'] = df.apply(calculate_full_time, axis=1)
df
```

<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>
        به کمک تابع <code>sort_values</code> سطرهای دیتافریم را بر اساس ستون full_time به ترتیب نزولی مرتب می‌کنیم.
        <br>
        سپس ۱۰ سطر اولیه دیتافریم مرتب‌شده را درون متغیر <code>most_time_consuming</code> می‌ریزیم.
</div>


```python
df_most_time_consuming = df.sort_values(by='full_time', ascending=False)[:10].drop('full_time', axis=1)
```

```python
import zipfile
import joblib

joblib.dump(spicy_ratio,"spicy_ratio")
df_most_time_consuming.to_csv('df_most_time_consuming.csv', index=False)


def compress(file_names):
    print("File Paths:")
    print(file_names)
    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED
    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = ["spicy_ratio", "df_most_time_consuming.csv", "hendoone.ipynb"]
compress(file_names)
```
